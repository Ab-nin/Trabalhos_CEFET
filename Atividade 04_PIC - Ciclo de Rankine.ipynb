### Eficiência de Carnot

$$\eta_c = 1 - \dfrac{T_l}{T_h} = 1 - \dfrac{T_1}{T_3}$$

```python
eta_c = (1 - (T1/T3))*100
display(Math(r'\eta_{c} = %.3f \, \%% ' %eta_c))
```

### Eficiência Térmica do Ciclo

$$\eta_t = \dfrac{h_3 - h_4 - [v_1\times(p_1 - p_2)]}{h_3 - h_2}$$

```python
eta_t = (((h3_c - h4_c) - (v_1 * ((p_l/1000) - (p_h/1000)))) / (h3_c - h2_c)) * 100
display(Math(r'\eta_{t} = %.3f \, \%% ' %eta_t))
```

```python
delta_eta = ((eta_c - eta_t)/eta_t)*100
display(Math(r'\text{A eficiência de Carnot é cerca de } %.2f \, \%% \text{ maior que a eficiência térmica real do ciclo.}' % delta_eta))
```
